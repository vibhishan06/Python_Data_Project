```python
# Installing datasets library
# !pip install datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_country']
```

Lets look at the top 20 countries on the dataframe:

```python
df['job_country'].value_counts().head(20)
```

Lets try to check if Brazil is in the dataframe:

```python
df['job_country'].isin(['Brazil']).any()
```

Lets check if North Korea is in the dataframe:

```python
df['job_country'].isin(['North Korea']).any()
```

Let's create a new dataframe with only our country of choice filtered for:

```python
us_jobs = df[df['job_country'] == "United States"]
```

```python
us_jobs
```

Removing all values where salary_year_avg is Nan:

```python
us_jobs = us_jobs[us_jobs['salary_year_avg'].notna()]

us_jobs
```

Count of jobs based on job_title_short:

```python
us_jobs.groupby('job_title_short').count()
```

Min salary grouping by job title

```python
us_jobs.groupby('job_title_short')[['salary_year_avg']].min()
```

Grouping by median, min, max and count

```python
us_jobs.groupby('job_title_short')[['salary_year_avg']].agg(['median', 'min', 'max', 'count'])
```

Sorting them by median and in descending order:

```python
us_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by = 'median', ascending = False)
```
