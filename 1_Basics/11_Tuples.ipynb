```python
lukes_skills = ("python", "sql", "statistics", "tableau")
print(id(lukes_skills))

new_skills = ("excel", "power bi")

lukes_skills += new_skills

lukes_skills                        # both are not the same in tuples
id(lukes_skills)
```

```python
lukes_skills = ["python", "sql", "statistics", "tableau"]
print(id(lukes_skills))

new_skills = ["excel", "power bi"]

lukes_skills += new_skills

lukes_skills                        # both are same in lists
id(lukes_skills)
```
