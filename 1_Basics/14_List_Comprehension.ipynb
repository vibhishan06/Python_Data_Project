**LIST COMPREHENSION**

Not limited to only list Comprehension:  
* set Comprehension  
* tuple Comprehension  
* dictionary Comprehension



```python
numbers = [x for x in range(100)]

numbers
```

```python
numbers = []

for number in range(100):
  numbers.append(number)
```

```python
numbers
```

```python
numbers = [x for x in range(100)]

numbers
```

```python
numbers = [x*2 for x in range(100)]                   # multiplies each number by 2

numbers
```

```python
numbers = [float(x) for x in range(100)]

numbers
```

```python
numbers = [x for x in "PYTHON"]                       # in case of strings

numbers
```

```python
numbers = {x for x in "PYTHON"}                       # gives out values in order

numbers
```
