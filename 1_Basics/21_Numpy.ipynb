```python
# Job titles
# Suppose we have a job title and we don't have any info on its base salary and bonus rate, we will use nan (not a number) in such cases

job_titles = np.array(["Data ANalyst", "data Scientist", "Data Enginner", "Machine Learning Engineer", "AI Engineer"])

# Base Salaries
base_salaries = np.array([60000, 80000, 75000, 90000, np.nan])

# Bonus rate
bonus_rates = np.array([.05, .1, .08, .12, np.nan])
```

```python
total_salaries = base_salaries * (1 + bonus_rates)

total_salaries
```

```python
np.nanmean(total_salaries)
```
