**LOGICAL OPERATORS**
---
and, or, not

```python
job_work_from_home = False
job_health_insurance = False
job_no_degree_mention = True

job_work_from_home and job_health_insurance                 # gives True only if all conditions are true.
```

```python
job_work_from_home = False
job_health_insurance = False
job_no_degree_mention = True

job_work_from_home or job_no_degree_mention                 # gives True if any one condition is true.
```

```python
not True
```

```python
if job_work_from_home or job_health_insurance or not job_no_degree_mention:
  print("Meets conditions")
```

Membership Operators


---
in, not in


```python
"Data" in "Data Nerd"                   #case-sensitive
```

```python
job_skills = ["python", "r", "sql"]
```

```python
"python" in job_skills                  #checking whether it contains a value
```

```python
"python" not in job_skills
```

Identity Operators

---
is, is not


```python
salary_kelly = 100_000
salary_luke = 100_000

salary_kelly is salary_luke
```

```python
id(salary_kelly)
```

```python
id(salary_luke)
```

```python
id(salary_kelly) == id(salary_luke)
```

```python
core_skills = ["Python", "R", "SQL"]
lukes_skills = core_skills
```

```python
lukes_skills is core_skills
```
