```python
# Applicant Skill
applicant_skill = "SQL"

# Skill required for Job posting
job_skill = "SQL"
```

```python
if job_skill == applicant_skill:
  print("Skills Matched!")
```

```python
job_skill = "SQL"
applicant_skill = "Python"
years_experience = 3

if job_skill == applicant_skill:
  print("Skills Matched!")
elif years_experience >= 5:                   # we can use multiple elif statements
  print("Enough Experience; No SQL")
elif applicant_skill == "Python":
  print("This applicant should know SQL")
else:
  print("No Skill or Experience")
```
