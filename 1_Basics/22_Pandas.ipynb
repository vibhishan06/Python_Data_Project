**PANDAS - ANALYZING**

```python
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.describe()
```

```python
df.info()
```

```python
df.count()
```

```python
df['salary_year_avg'].mean()
```

```python
df['salary_year_avg'].median()
```

Lets find the minimum value in salary_year_avg

```python
df['salary_year_avg'].min()
```

Position of the minimum salary in the dataframe:

```python
min_salary = df['salary_year_avg'].idxmin()               # .iloc stands for "integer location" - it lets you select rows and columns from a DataFrame using their position numbers (like index numbers in a list).
```

```python
df.iloc[min_salary]
```

Getting the unique values from the job_title_short column

```python
df['job_title_short'].unique()
```

Now, lets find the unique value counts of each of these titlies

```python
df['job_title_short'].value_counts()
```

Now, let's find out what is the minimum salary in each of the job titles above:

```python
df.groupby('job_title_short')['salary_year_avg'].min()
```

Median salary:

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

Maximum salary:

```python
df.groupby('job_title_short')['salary_year_avg'].max()
```

Performing aggregation on multiple columns:

```python
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median()
```

Getting min, max and median alltogether:

```python
df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max', 'median'])
```
