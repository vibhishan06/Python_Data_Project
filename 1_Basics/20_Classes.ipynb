```python
class LukesList:
  def __init__(self):
    """Initialize an empty list"""
    self.items = []

  def add(self, item):
    """Add an item to the end of the list"""
    self.items.append(item)

  def __getitem__(self, index):
    """Retrieve an item by index"""
    return self.items[index]

  def __setitem__(self, index, value):
    """Set an item at a specific index"""
    self.items[index] = value

  def __repr__(self):
    """Return a string representation of the list"""
    return str(self.items)

  def __len__(self):
    """Return the length of the list"""
    return len(self.items)

```

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=0.1, symbol="$", ):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol

  def __repr__(self):
    return f"{self.symbol}{self.base_salary:,.0f}"

  def calculate_total_salary(self):
    total_salary = self.base_salary * (1 + self.bonus_rate)
    return f"{self.symbol}{total_salary:,.0f}"
```

```python
data_salary = BaseSalary(100000, .2)
```

```python
data_salary
```

```python
data_salary.calculate_total_salary()
```

Another way to do this is:

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=0.1, symbol="$"):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total_salary = base_salary * (1 + bonus_rate)
    self.bonus = self.total_salary - self.base_salary

  def __repr__(self):
    return f"{self.symbol}{self.total_salary:,.0f}"

  def show_salary(self):
    return f"{self.symbol}{self.total_salary:,.0f}"

  def show_bonus(self):
    return f"{self.symbol}{self.bonus:,.0f}"
```

```python
salary = BaseSalary(100000)

salary.symbol
```

```python
salary
```

```python
salary.show_salary()
```

```python
salary.show_bonus()
```
