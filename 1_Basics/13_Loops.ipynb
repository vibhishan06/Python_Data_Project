for Loops

```python
numbers = [0, 1, 2, 3, 4, 5]

for number in numbers:
  print(number)
```

```python
characters = "PYTHON"

for character in characters:
  print(character)
```

```python
years_experience = {
    "Luke": 3,
    "Kelly": 6,
    "Ken": 4,
    "ALex": 3
}
```

```python
for person in years_experience:             # Gives the keys
  print(person)
```

```python
for value in years_experience.values():
  print(value)
```

```python
for key, value in years_experience.items():
  print(key, value)
```

```python
for key, value in years_experience.items():
  if value > 5:
    print(key, "has", value, "years of experience")
```

```python
range(5)
```

```python
for i in range(100):
  print(i)
```

while Loops

```python
count = 1
while count <= 10:
  print(count)
  count += 1
```

```python
years_experience = {
    "Luke": 3,
    "Kelly": 6,
    "Ken": 4,
    "ALex": 3,
    "Mary": 7
}
```

```python
index = 0
years_list = list(years_experience.items())

while index < len(years_list):
    key, years = years_list[index]
    if years >= 5:
        print(key, "has", years, "years of experience.")
    index += 1
```
