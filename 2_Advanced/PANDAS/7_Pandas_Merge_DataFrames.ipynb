```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

In the last file, we made a dataframe which analyzed jobs in United States, specific to different job titles and how the counts of those different job titles trend over the different months over the year [file_no.6, code cell: 36]. We are going to merge that data with a new data of software_jobs....

We are using the read_csv function from pandas to extract the software_jobs data.

Bringing back the df_US_pivot dataframe........

```python
df_US = df[df['job_country'] == 'United States'].copy()
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_US_pivot = df_US_pivot.reset_index()
df_US_pivot['job_posted_month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot = df_US_pivot.sort_values('job_posted_month_no')
df_US_pivot = df_US_pivot.set_index('job_posted_month')
df_US_pivot = df_US_pivot.drop(columns='job_posted_month_no')

df_US_pivot
```

```python
df_US_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col = 'job_posted_month')

df_US_software_pivot
```

```python
df_US_merged = df_US_pivot.merge(df_US_software_pivot, on = 'job_posted_month')
```

We want to now plot the top 5 jobs based on count of jobs in the year and see how they trend over time.........

Creating a list of top 5 jobs, then filtering the dataframe to include those, which makes it easier to plot.

```python
top_5 = (df_US_merged
        .sum()
        .sort_values(ascending = False)
        .head(5)
        .index
        .to_list()
)

top_5
```

```python
df_US_merged[top_5].plot(kind = 'line')
```

Cleaning it.........

```python
df_US_merged[top_5].plot(kind = 'line')
plt.title('Monthly Job Postings for Top Tech Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 22000)
plt.legend()
plt.show()
```
