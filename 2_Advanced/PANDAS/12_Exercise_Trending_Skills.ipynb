**EXCERCISE-** Investigate Trending Skills

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

For this exercise, we will be focusing on Data Analysts, 

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_DA
```

Now, we will aggregate the skills on a monthly basis and for that we need to extract the 'month' from job_posted_date

```python
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month

df_DA
```

Now, lets get into exploding out the job_skills...

```python
df_DA_explode = df_DA.explode('job_skills')

df_DA_explode
```

So, now we are going to use both these columns (job_skills and job_posted_month_no) and we will pivot them

```python
df_DA_pivot = df_DA_explode.pivot_table(index = 'job_posted_month_no', columns = 'job_skills', aggfunc = 'size', fill_value = 0)
```

```python
df_DA_pivot
```

Let's create a row called Total that sums up all the values and sort these columns on the basis of the total....

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending = False).index]
```

Now, dropping the Total row...

```python
df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot
```

Now its PLOTTING TIME!!!!!!!!!!!

```python
df_DA_pivot.iloc[:, :5].plot(kind = 'line')

plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

Converting those numbers on the x-axis into names.....

```python
df_DA_pivot = df_DA_pivot.reset_index()
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')

df_DA_pivot.iloc[:, :5].plot(kind = 'line')

plt.title('Top 5 Skills for Data Analysts per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```
